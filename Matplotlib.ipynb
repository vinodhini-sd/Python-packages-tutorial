```
x=np.linspace(0,0.5,15)
y = [x**2 for x in x]
plt.plot(x,x,'bo',label = 'linear')
#plt.show()  # plots one graph, shows it, so next plt.plot will create a new graph
# plt.scatter(x,y,'ro',label = 'poly') gives error because the 3rd argument is not the color but the size of the points. thus the arguments order error. refer to scatter documentation
plt.scatter(x,y,c = 'r',label = 'poly')
plt.legend()
plt.show()
```

# **Sub plots in a figure**

```
fig = plt.figure()
ax1 = fig.add_subplot(211) #coordinates of all the subplots are not in our control
ax1.plot(x,x)
ax2 = fig.add_subplot(212) #212 - (#rows, #cols, #plot number)
ax2.scatter(x,y)
plt.show()
```

```
fig = plt.figure()
fig.add_axes([0.1,0.1,0.8,0.8])  #so the plot area has dimensions of 0 to 1 in x and y directions both #perfect coordinates can be selected for the plot
#plt.barh(y, x) #this is blocking the other graphs underneath how to change opaque transparency, etc
plt.plot(x,x)
plt.scatter(x,y)
plt.show()
```

```
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.plot(x,y)
ax2.hist(x)
plt.show()
```

```
fig, ax = plt.subplots()
ax.plot(df.Date, df.Close, color = 'r', marker = 'o', markerfacecolor = 'r', label = 'Close')
ax.set(ylim = [min(df.Close), round(max(df.Close))], xlim = [min(df.Date), max(df.Date)], xlabel = 'Date', ylabel = 'Closing Value', title = 'Closing stock value of Alphabet Inc')
ax.set(xticklabels = df.Date)
ax.legend()
ax.minorticks_on()  #only when minorticks arw on, we can plot minor grids. otherwise, no.
ax.grid(which = 'major', axis = 'both', linestyle = '--', linewidth = 0.7, color = 'gray')
ax.grid(which = 'minor', linestyle = 'dotted', linewidth = 1.3, color = 'yellow')
plt.show()
```

Grid needs to be written in another ax.grid() function call and not part of ax.set().

Only when the ticks are on, we can plot the grids. so major ticks should be on for major grids and minor ticks on for minor grids.

ax.tickparams() can decide which axis, major or minor and direction of tick, in or out or anything else, can be set.

```
fig, ax = plt.subplots()
ax.plot(df.Date, df.Close, color = 'r', marker = 'o', markerfacecolor = 'r', label = 'Close')
ax.set(ylim = [min(df.Close), round(max(df.Close))], xlim = [min(df.Date), max(df.Date)], xlabel = 'Date', ylabel = 'Closing Value', title = 'Closing stock value of Alphabet Inc')
ax.set(xticklabels = df.Date)
ax.legend()
ax.minorticks_on()  #only when minorticks arw on, we can plot minor grids. otherwise, no.
ax.grid(which = 'both')
ax.tick_params(which = 'both', direction = 'in')
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(index = None)
df['Programming languages'] = ['Jave', 'Pythin', 'PHP', 'Javascript', 'C#', 'C++']
df['Popularity'] = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x = df['Programming languages']
y = df['Popularity']
```

```
fig, ax = plt.subplots()
ax.bar(x,y, width = 0.6, color = 'yellow', edgecolor = 'green', linewidth = 2.9, yerr = [1.2,1.3,1.5,1.2,1.4,1.2], ecolor = 'blue', align='edge')
ax.set(xlabel = "LAnguages", ylabel = "Popularity", title = 'Popularity of programming languages', ylim = [0,25])
ax.minorticks_on()
ax.grid(which = 'minor', color = 'g', linestyle = 'dotted', linewidth = 0.5)
ax.grid(which = 'major', color = 'r', linewidth = 0.7, linestyle = 'solid')
ax.tick_params(direction = 'in')
plt.show()
```

```
fig, ax = plt.subplots()
ax.barh(x,y, height = 0.7, align = 'center', color = 'yellow', edgecolor = 'green', linewidth = 2.9, xerr = [1.2,1.3,1.5,1.2,1.4,1.2], ecolor = 'blue')
ax.set(xlabel = "LAnguages", ylabel = "Popularity", title = 'Popularity of programming languages', xlim = [0,25])
ax.minorticks_on()
ax.grid(which = 'minor', color = 'g', linestyle = 'dotted', linewidth = 0.5)
ax.grid(which = 'major', color = 'r', linewidth = 0.7, linestyle = 'solid')
ax.tick_params(direction = 'in')
plt.show()
```
